```r
colnames(IMDB) <- c("Film","Genre","CriticRating","AudienceRating","BudgetInMillions","Year")
```

```r
head(IMDB)
tail(IMDB)
```

```r
str(IMDB)
summary(IMDB)
```

```r
IMDB$Year <- factor(IMDB$Year)
```

```r
library(ggplot2)
```

```r
ggplot(data = IMDB, aes(x = CriticRating, y = AudienceRating, colour = Genre, size = BudgetInMillions )) + geom_point()
```

```r
ggplot(data = IMDB, aes(x = BudgetInMillions, y = AudienceRating, colour = Genre )) + geom_point()
```

```r
ggplot(data = IMDB, aes(x = BudgetInMillions)) + geom_histogram(binwidth =10, aes(fill = Genre), colour = "black")
```

```r
ggplot(data = IMDB) + geom_histogram(binwidth =10, aes(x = AudienceRating, fill = Genre), colour = "black")
```

```r
ggplot(data = IMDB) + geom_histogram(binwidth =10, aes(x = CriticRating, fill = Genre), colour = "black")
```

```r
ggplot(data = IMDB, aes(x=CriticRating, y= AudienceRating, colour = Genre )) + geom_smooth(fill = NA)
```

```r
ggplot(data = IMDB, aes(x=Genre, y= AudienceRating, colour = Genre )) + 
geom_jitter() + geom_boxplot(alpha = 0.5)
```

```r
ggplot(data = IMDB, aes(x=Genre, y= CriticRating, colour = Genre )) + 
geom_jitter() + geom_boxplot(alpha = 0.5)
```

```r
ggplot(data=IMDB,aes(x=BudgetInMillions)) + geom_histogram(binwidth=10,color='black',aes(fill = Genre))+
  facet_grid(Genre~.,scales = 'free')
```

```r
ggplot(data = IMDB, aes(x= AudienceRating, y = CriticRating, colour = Genre)) +
geom_point() + facet_grid(Genre~.)
```

```r
ggplot(data = IMDB, aes(x= CriticRating , y =AudienceRating , colour = Genre)) +
geom_point() + geom_smooth()  + facet_grid(Genre~Year) + coord_cartesian(ylim=c(0,100))
```

```r
ggplot(data = IMDB, aes(x = BudgetInMillions)) + 
xlab("Budget") + ylab("Number of movies") + ggtitle("Movie and Budget Relation") + 
geom_histogram(binwidth =10, aes(fill = Genre), colour = "black") +
theme(legend.position = c(1,1),
     legend.justification= c(1,1),
     plot.title= element_text(colour = "Blue", size = 30, family = "courier" ))
```
